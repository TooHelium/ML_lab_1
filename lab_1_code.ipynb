```python
import numpy as np
import matplotlib as plt
```

```python
#Візуалізувати дані з lab1_train.csv.
file_path = "lab_1_train.csv"
data = np.loadtxt(file_path, delimiter=',', skiprows=1)
    
x = data[:, 1]
y = data[:, 2]

plt.plot(x, y)
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

```python
#Після кожної епохи виводити поточні значення ваг та функції втрат. 
#Використовувати Mean Squared Error як функцію втрат.
file_path = "lab_1_train.csv"
data = np.loadtxt(file_path, delimiter=',', skiprows=1)
x_train = data[:, 1]
y_train = data[:, 2]

m = 0
c = 0

learning_rate = 0.001  
epochs = 10000

n = float(len(x_train))

for i in range(epochs): 
    Y_pred = m * x_train + c
    D_m = (-2/n) * sum(x_train * (y_train - Y_pred))
    D_c = (-2/n) * sum(y_train - Y_pred)
    m = m - learning_rate * D_m
    c = c - learning_rate * D_c
    
    loss = np.mean((y_train - Y_pred)**2)
    
    #print(f"Epoch {i+1}: m = {m}, c = {c}, loss = {loss}")
```

```python
#Після того, як градієнтний спуск зійшовся, порахувати функцію втрат на тестових даних з lab1_test.csv.
file_path = "lab_1_test.csv"
data = np.loadtxt(file_path, delimiter=',', skiprows=1)
x_test = data[:, 1]
y_test = data[:, 2]

y_pred_test = m * x_test + c

loss = np.mean((y_test - y_pred_test)**2)

print(loss)
```

```python
#Візуалізувати на одному графіку дані з lab1_train.csv,  lab1_test.csv та лінію, до якої зійшовся розв’язок.
train_data = np.loadtxt("lab_1_train.csv", delimiter=",", skiprows=1)
x_train = train_data[:, 1]
y_train = train_data[:, 2]

test_data = np.loadtxt("lab_1_test.csv", delimiter=",", skiprows=1)
x_test = test_data[:, 1]
y_test = test_data[:, 2]

plt.scatter(x_train, y_train, color='yellow')
plt.scatter(x_test, y_test, color='blue')

line = m * x_train + c
plt.plot(x_train, line, color='red')

plt.xlabel('x')
plt.ylabel('y')
plt.show()
```
